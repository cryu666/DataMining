```python
import pandas as pd

data = pd.read_csv("D:/Data Mining/HW3/新竹_2021.csv")
data.info()
```

```python
import datetime as dt

data = data.drop(0)   #刪除虛線列
data["日期"] = data["日期"].apply(lambda x:dt.datetime.strptime(x.split(" ")[0], "%Y/%m/%d").timestamp())
data = data[data["日期"].apply(lambda x: x >= dt.datetime(2021, 10, 1).timestamp() and x < dt.datetime(2022, 1, 1).timestamp())]

data.head()
```

```python
for i in range(24):
    data[str(i)] = data[str(i)].apply(lambda x: 0 if x == "NR" else x)
```

```python
import numpy as np

data = data.T
data = data.iloc[2:]   #刪除測站列、日期列
test_item_list = list(data.iloc[0,:])[:18]
for i in range(len(test_item_list)):
    test_item_list[i] = test_item_list[i].split(" ")[0]

split_data = []
for i in np.arange(0, len(data.T), 18):
    element = data.iloc[1:, i:i + 18]
    element.columns = test_item_list
    split_data.append(element)

timeseries_data = pd.concat(split_data, ignore_index = True)
timeseries_data = timeseries_data.set_index(pd.date_range(start = "2021/10/01", end = "2022/1/1", freq = "H")[:2208])
timeseries_data
    

```

```python
invalid = ["#", "*", "x", "A"]
for i in range(len(timeseries_data.columns)):
    col_list = list(timeseries_data.iloc[:, i])
    for j in range(len(col_list)):
        if any(invalid_char in col_list[j] for invalid_char in invalid):
            try:
                last_element = float(col_list[j - 1])
            except:
                last_count = 1
                while(any(invalid_char in str(col_list[j - last_count]) for invalid_char in invalid)):
                    last_count += 1
                    try:
                        last_element = col_list[j - last_count]
                    except:
                        last_element = 0.0
                last_element = float(last_element)
                
            try:
                next_element = float(col_list[j + 1])
            except:
                next_count = 1
                while(any(invalid_char in str(col_list[j + next_count]) for invalid_char in invalid)):
                    next_count += 1
                    try:
                        next_element = col_list[j + next_count]
                    except:
                        next_element = 0.0
                next_element = float(next_element)
            
            col_list[j] = last_element + next_element / 2
            
    timeseries_data[timeseries_data.columns[i]] = col_list
    
timeseries_data = timeseries_data.astype("float64")
```

```python
train = timeseries_data.loc[:"2021-11-30 23:00:00", :]
test = timeseries_data.loc["2021-12-01 00:00:00":, :]
```

```python
def getFeatureTarget(data, feature_num, predict_hour):
    X = []
    Y = []
    for i in range(len(data) - (6 + predict_hour)):
        if feature_num == 1:
            X.append(data.iloc[i: i + 6, 9])
        else:
            X.append(data.iloc[i: i + 6, :].values.tolist())
        Y.append(data.iat[i + 5 + predict_hour, 9])
            
    return np.array(X), np.array(Y)
```

```python
from sklearn.linear_model import LinearRegression
import xgboost as xgb
from sklearn.metrics import mean_absolute_error

train_X, train_Y = getFeatureTarget(train, 1, 1)
test_X, test_Y = getFeatureTarget(test, 1, 1)

model_LR_1_1 = LinearRegression()
model_LR_1_1.fit(train_X, train_Y)
print('The MAE loss of testing set(Linear Regression): {:.4f}'
      .format(mean_absolute_error(test_Y, model_LR_1_1.predict(test_X))))

model_XGB_1_1 = xgb.XGBRegressor()
model_XGB_1_1.fit(train_X, train_Y)
print('The MAE loss of testing set(XGBoost): {:.4f}'
      .format(mean_absolute_error(test_Y, model_XGB_1_1.predict(test_X))))
```
